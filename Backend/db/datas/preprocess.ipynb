```python
import pandas as pd
```

```python
books = pd.read_csv('Books.csv', encoding='cp949')
```

```python
books.rename(columns={
                'ISBN': 'isbn',
                'Book-Title': 'title',
                'Book-Author': 'author',
                'Publisher': 'publisher',
                'Year-Of-Publication': 'publication_year',
                'Image-URL-S': 'image_S_URL',
                'Image-URL-M': 'image_M_URL',
                'Image-URL-L': 'image_L_URL',
                }, inplace=True)
```

```python
books['isbn_str'] = books.apply(
                    lambda row: row.image_S_URL.split('/')[5].split('.')[0]
                    ,axis=1)

books.drop_duplicates(['isbn_str'], keep='first', ignore_index=True, inplace=True)
```

```python
b_isbn = set(books.isbn.values)
b_isbn_str = set(books.isbn_str.values)
```

```python
ratings = pd.read_csv('Ratings.csv', encoding='cp949')
ratings.rename(columns={
                'User-ID': 'user_id',
                'ISBN': 'isbn',
                'Book-Rating': 'rating'
                }, inplace=True)
```

```python
r_isbn = set(ratings.isbn.values)
```

```python
len(r_isbn - b_isbn)
```

```python
len(r_isbn - b_isbn_str)
```

```python
len((r_isbn & b_isbn) - b_isbn_str)
```

```python
only_r_isbn = r_isbn - (b_isbn | b_isbn_str)
len(only_r_isbn)
```
